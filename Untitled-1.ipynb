Accessing Info from a list into .csv format⬇️

```python
olympians=[("John Aalberg", 31, "Cross Country Skiing, 15KM"),
           
("Minna Maarit Aalto", 30, "Sailing"),

("Win Valdemar Aaltonen", 54, "Art Competitions"),

("Wakako Abe", 18, "Cycling"),]
```

```python
with open('olympic_info.txt', 'w') as file:
    file.write('Name, Age, Game\n')

    for olympian in olympians:
        line_string=('''{}:{},{}'''.format(olympian[0], olympian[1], olympian[2]))
        file.write(line_string)
```

```python
with open('olympic_info.txt', 'r') as now:
    seeya=now.readlines()
    print(seeya)
```
