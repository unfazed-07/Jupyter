### Stack


```python
class Stack:
    def __init__(self):
        self.values = []
    def push(self, x):
        self.values.append(x)
    def pop(self):
        return self.values.pop()

s = Stack()
s.push(10)
s.push(20)
s.push(30)
s.push(45)
print(s.values)
print(s.pop())
print(s.values)
```

```python
stack = []

n = int(input())

for i in range(n):
    x = int(input())
    stack.append(x)

res = min(stack)
print(res)
```

### Linked list

```python
class Node:
    def __init__(self, data=None, next = None):
        self.data = data
        self.next = None
    
a = Node(5)
b = Node(3)
c = Node(7)

a.next = b
b.next = c

head = a
print(head.data)
print(a.next.data)
print(a.next.next.data)
```
