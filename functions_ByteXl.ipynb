Lambda -> Small Anonymous (nameless) function witten in a single line.

Syntax:
Lambda parameter : Expression

### Properties
- No Def keyword
- No function name
- No return function
- Only one expression
- Cant take multiple arguments
- Automatically return 

```python
#Normal User Defined function
def square(x):
    return x*x
print(square(4))
```

```python
# Lambda Function
square = lambda x: x*x 
print(square(4))
```

```python
find = lambda x : "Even" if x%2==0 else "Odd"
print(find(21))
```

```python
add = lambda *a: sum(a)
print(add(2,4,1))
```

```python
maxx = lambda a, b : a if a>b else b
print(maxx(10, 4))
```

```python
#Lambda with Map
num = [11,22,33,44,55]
even = set(map(lambda x: x*x, num))
even
```

```python
num = [1,2,3,4,5,6]
evens = list(filter(lambda x: x%2 == 0, num))
evens
```

```python
students = [("Riya", 90), ("Ankit", 82), ("Neha", 92), ]
sorted_students = sorted(students, key= lambda x: x[1])
sorted_students
```

```python
data = {"a":5, "b": 10, "c": 21}
sorted_dict = sorted(data.items(), key = lambda x: x[1])
sorted_dict
```

Recursion: A meathod calling itself
Must have:
- Base Call
- Revursive Call

```python
def factorial(n):
    if n==1:
        return 1
    return n*factorial(n-1)

print(factorial(5))
```

```python
#Fibonacci Series
def fibonacci(n):
    if n<=1:
        return n  
    return fibonacci(n-1)+fibonacci(n-2)

for i in range(6):
    print(fibonacci(i))
```

```python
# Reverse a string using recursion
def reverse(str):

    if len(str)==0:
        return ""
    return str[-1] + reverse(str[:-1])

str = "Freak"
print(reverse(str))
```
