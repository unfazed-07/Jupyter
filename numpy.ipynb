YOu are given marks of students in three subjects, you need to perfrom, calculate avg marks, using numpy, store data in pandas df, find topper, plot avg marks comparison graph

```python
import numpy as np
name = ["Name", "Riya", "Ansh"]
marks = np.array([
    [80,73,65],
    [68, 67, 88],
    [44, 23, 36]
])

avg = np.mean(marks, axis = 1)
df = pd.DataFrame(marks, columns=["Math", "Science", "English"])
df["name"] = name
df["Average"] = avg
topper = df.loc[df["Average"].idxmax()]
topper
```

```python
plt.bar(df["name"], df["Average"])
plt.title("Average Marks")
```

```python
# Sales Data Analysis Dashboard

months = ["Jan", "Feb", "Mar", "Apr"]
sales = [250, 300, 250, 400]
#Find total and average sales
# Idenfify the hgihest sales month
sales = np.array(sales)

df = pd.DataFrame({"Months": months,
                   "Sales": sales})

print("Total:", np.sum(sales))
print("Average:", np.average(sales))
print("Higest Sales Month:", df.loc[df["Sales"].idxmax()]["Months"])
```
