### Stack


```python
class Stack:
    def __init__(self):
        self.values = []
    def push(self, x):
        self.values.append(x)
    def pop(self):
        return self.values.pop()

s = Stack()
s.push(10)
s.push(20)
s.push(30)
s.push(45)
print(s.values)
print(s.pop())
print(s.values)
```

```python
stack = []

n = int(input())

for i in range(n):
    x = int(input())
    stack.append(x)

res = min(stack)
print(res)
```

### Linked list

```python
class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None
    
a = Node(5)
b = Node(3)
c = Node(7)

a.next = b
b.next = c

head = a
print(head.data)
print(a.next.data)
print(a.next.next.data)
```

```python
#traverse linked list ✅
def printLinkedList(head):
    curr= head
    while (curr != None):
        print(curr.data, end = " ")
        curr = curr.next
printLinkedList(head)
```

```python
#insertion at beginning
newNode = Node(4)
newNode.next = head
head = newNode
printLinkedList(head)
```

```python
#insertion at the end
newNode = Node(1)
curr = head
while curr.next != None:
    curr = curr.next

curr.next = newNode

printLinkedList(head)
```

```python
# insertion at kth index
k = 2
newNode = Node(6)
curr = head
for i in range(k-1):
    curr = curr.next

newNode.next = curr.next
curr.next = newNode
printLinkedList(head)
```

### Doubly Linkedin List

```python

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None

n = int(input())

head = None
tail = None

for i in range(n):
    val =int(input())
    newNode = Node(val)


    if head is None:
        head = newNode
        tail = newNode
    else:
        tail.next = newNode
        newNode.prev = tail
        tail = newNode

curr = head

while curr is not None:
    if curr.next is not None:
        print(f"{curr.data} -> ", end = "")
    else:
        print(f"{curr.data}")

    curr = curr.next


curr = tail
while curr is not None:
    print(f"{curr.data} <->", end = "")
    curr = curr.prev
```
