Pattern  Meaning
\d  Digit
\D Non-Digit
\w Word(a-z, A-Z, 0-9)
\W Non-Word
\s Space
\S Non Spaces
+  One or more
*  Astrik
{n.m} Range


```python
import re
text = "hello World"
pattern = "hello"
match = re.search(pattern, text)

if match:
    print("Match found:", match.group())
```

```python
text = "hello h2llo h3llo"
pattern = "h.llo"
match = re.search(pattern, text)

print(match)
```

```python
text = "abcdef"
print(re.search("^abc", text))
print(re.search("def$", text))
```

```python
text = "hello"
pattern = "[aeiou]"
print(re.findall(pattern , text))
```

```python
text = "aaab"
pattern = "a{2,4}"
print(re.findall(pattern, text))
```

```python
email = "test123@gmail.com"
pattern= r"^[a-zA-Z0-9._]+@[a-z]+\.[a-z]{2,}$"
if re.match(pattern, email):
    print("Valid Email")
else: print("Invalid Email")
```
