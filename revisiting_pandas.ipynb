```python
import pandas as pd 
data = {
    "Region": ["N", "N", "N", "S", "S", "S"],
    "Category": ["A","A","B","B","B","A"],
    "Sales": [100,200,350,250,150,200]
}
df = pd.DataFrame(data)

df.groupby(["Region"]).agg(avg_sales = ("Sales","mean"), total_revenue = ("Sales","sum"))
```
